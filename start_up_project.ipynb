```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/content/drive/MyDrive/R K Data Analytic/Colab Notebooks/Data Set/startup_funding (1).csv")
df
```

```python
#to delete the column
df.drop(columns = ['Remarks'], inplace = True)
df
```

```python
#which col is present in the data set withe there datatype
df.info()
```

```python
#to remane the col name
df.rename(columns = {"Date dd/mm/yyyy":"Date",
                    "Startup Name": "Startup",
                     "SubVertical":"Subvertical",
                     "City  Location": "City",
                     "Amount in USD":"Amount"}, inplace = True)
df
```

- in Amount column ',' in present hence it show as string, for that we have to replace the ',' with nothing('')
- also replace undisclosed with nothing('')

```python
# if we perform the operation we have to specify witch col we have to modify in front of operation.
#col name   =  replace
df['Amount'] = df['Amount'].str.replace(',', '')
```

- Amount datatype is object we have to convert it into integer or float

```python
#pd.numeric is used to convert the passed argument to a numeric type.
#errors = 'coerec'- convert problematic inputs to NaT (Not a Time).
df['Amount'] = pd.to_numeric(df['Amount'] ,errors='coerce')
```

```python
df['Date']
```

```python
#convert date formate
df['Date']= pd.to_datetime(df['Date'],errors ='coerce' ,format="%d/%m/%Y")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#to find the percentage
df.isnull().mean() * 100
```

```python
# fillna - replaces the NULL values with a specified value
df['Amount'] = df['Amount'].fillna(df['Amount'].mean())
```

```python
df.head()
```

```python
#avg of amount col
df['Amount'].mean()
```

```python
df.isnull().mean() * 100
```

```python
#To calculate the mode of a group of numbers
df['Subvertical'].mode()
```

```python
#Fill the missing values in subvertical column with unknown
df['Subvertical'] = df['Subvertical'].fillna('Unknown')
```

```python
df['Year'] = df['Date'].dt.year
```

```python
df.head()
```

```python
#removes the rows that contains NULL values
df.dropna(subset=['Date','Industry Vertical','City','Investors Name','InvestmentnType'],inplace = True)
df
```

```python
#find the start up in which Ratan tata invested
df[df['Investors Name'] == 'Ratan Tata']
```

```python
df[df['Investors Name'] == 'Ratan Tata'].head(5)
```

```python
#1. How has the total funding amount changed over the years?
df.groupby('Amount')['Year'].sum()
```

```python
#2. Which are the top 5 most-funded startups?
df.groupby('Startup')['Amount'].sum().sort_values(ascending=False).head(5)
```

```python
#3. Which sub-sectors receive the highest funding?
df.groupby('Subvertical')['Amount'].sum().sort_values(ascending=False)
```

```python
#4. Which investors have repeatedly funded specific startups, and how frequently?
fund = df.groupby(['Investors Name', 'Startup']).size().reset_index(name='Funding Count')
fund[fund['Funding Count'] > 1]
```

```python
#5. What is the total funding amount for each city where "IDG Ventures" was an investor?
inv =  df[df['Investors Name'].str.contains("IDG Ventures", na=False)]
inv.groupby(['City', 'Investors Name'])['Amount'].sum().reset_index(name='Total Amount')
```

```python
#6. What is the amount spent by the investor in each startup?
df.groupby(['Investors Name', 'Startup'])['Amount'].sum()
```

```python
#7. Who are the top investors in the FinTech sector, and how much have they invested?
fintech_investments = df[df['Industry Vertical'] == 'FinTech']
top_investors_fintech = fintech_investments.groupby(['Investors Name'])['Amount'].sum().sort_values(ascending=False)
top_investors_fintech
```

```python
#8. What is the average funding amount?
df['Amount'].mean()
```

```python
# 1. Funding Over Time
funding_by_year = df.groupby('Year')['Amount'].sum().reset_index()
fig1 = px.line(funding_by_year, x='Year', y='Amount', title='Total Funding Over the Years',
               labels={'Amount': 'Total Funding (INR)', 'Year': 'Year'})
fig1.show()
```

```python
# 2. Top Startups by Funding
top_startups = df.groupby('Startup')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False).head(10)
fig2 = px.bar(top_startups, x='Startup', y='Amount', title='Top 10 Funded Startups',
              labels={'Amount': 'Total Funding (INR)', 'Startup': 'Startup'})
fig2.show()
```

```python
# 3. Funding Distribution by Industry
industry_funding = df.groupby('Industry Vertical')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False).head(10)
fig3 = px.pie(industry_funding, names='Industry Vertical', values='Amount', title='Funding Distribution by Industry')
fig3.show()
```

```python
# 4. Funding by City
city_funding = df.groupby('City')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False).head(10)
fig4 = px.bar(city_funding, x='City', y='Amount', title='Top 10 Cities by Funding',
              labels={'Amount': 'Total Funding (INR)', 'City': 'City'})
fig4.show()
```
